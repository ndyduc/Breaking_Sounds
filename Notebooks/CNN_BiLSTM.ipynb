```python
import glob
import os
import torch
import torch.optim as optim
import time
from Model.CNN_BiLSTM import * 
import pandas as pd
import time


shuffle_file = "../Trained/schuffle.txt"
df = pd.read_csv(shuffle_file, header=None, names=["wav_path", "csv_path"])

# Thêm cột index để theo dõi file nào đang được train
df.reset_index(inplace=True)
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")

model = CNN_BiLSTM().to(device) 
criterion = torch.nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

checkpoint_path = "../Trained/CNN_BiLSTM11.pth"
if os.path.exists(checkpoint_path):
    checkpoint = torch.load(checkpoint_path, map_location=device)
    model.load_state_dict(checkpoint["model_state_dict"])
    optimizer.load_state_dict(checkpoint["optimizer_state_dict"])
    
    
test_wav_files = glob.glob("../Data/MusicNet_Dataset/musicnet/musicnet/test_data/*.wav")
test_csv_files = glob.glob("../Data/MusicNet_Dataset/musicnet/musicnet/test_labels/*.csv")

for i, row in df.iterrows():
    index = row["index"]
    wav_path = row["wav_path"]
    csv_path = row["csv_path"]
    
    print(time.strftime("%H:%M:%S"), f"Training: {index}: {os.path.basename(wav_path)}")
    train_on(model, optimizer, criterion, wav_path, csv_path,test_wav_files, test_csv_files)

print("Training complete!")
```

```python
import pandas as pd
import glob
import os

folder_path = "../Data/MusicNet_Dataset/musicnet/musicnet/train_labels"

# Tìm tất cả các file CSV trong thư mục
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Tạo một tập hợp để lưu các giá trị duy nhất
unique_values = set()

for file in csv_files:
    df = pd.read_csv(file)
    if 'note_value' in df.columns:  # Kiểm tra xem cột có tồn tại không
        unique_values.update(df['instrument'].dropna().unique())

# In ra danh sách các giá trị duy nhất
print(sorted(unique_values))
```
